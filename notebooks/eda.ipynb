```python
import matplotlib.pyplot as plt

top_cities = df.sort_values(by="2023_accidents", ascending=False).head(10)

plt.figure()
plt.bar(top_cities['city'], top_cities['2023_accidents'])
plt.xticks(rotation=45)
plt.title("Top 10 Cities by Accidents (2023)")
plt.show()
```

```python
plt.figure()
plt.scatter(df['2023_accidents'], df['severity_ratio'])
plt.xlabel("Accidents")
plt.ylabel("Severity Ratio")
plt.title("Accidents vs Severity")
plt.show()
```
